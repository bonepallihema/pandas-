# Data Filtering and conditional changes

```python
f=pd.read_excel("panda.excel.xlsx")
f
```

```python
df=pd.DataFrame(f)
df
```

```python
df.loc[df["math"]>50]
```

```python
df.loc[df["science"]==90]
```

```python
df.loc[(df["math"]>60)&(df["telugu"]>50)]
```

```python
df.loc[(df["math"]>60)&(df["telugu"]>50)]
```

```python
df.loc[(df["math"]>60)&(df["telugu"]<90)]
```

```python
df.loc[df["stu name"].str.contains("n")]
```

```python
df.loc[df["stu name"].str.contains("h")]
```

```python
df["total"]=df["telugu"]+df["hindi"]+df["science"]+df["math"]
df["total"]
```

```python
df["perc"]=(df["total"]/400)*100
df
```

```python
df["grade"]="pass/fail"
df
```

```python
n=print(df.loc[df["perc"]<40,["grade"]]=="fail")
n
```

```python
df.loc[(df["perc"]>=40)&(df["perc"]<60),["grade"],]=="pass"
```

```python
df.loc[(df["perc"]>=60)&(df["perc"]<70),["grade"],]=="pass"
```

```python
df.loc[df["perc"]>=70,["grade"]]=="excellent"
```
